# EC2

## Swap

You should [always enable swap](http://unix.stackexchange.com/questions/190398/do-i-need-swap-space-if-i-have-more-than-enough-amount-of-ram) as it makes more efficient use of RAM. Swap is already on your mac, but does not automatically come with an EC2 instance.

To check if swap is enabled:
```
swapon -s
```

If it doesn't return anything, it is not enabled. 

It doesn't matter which directory you are in when you enable swap, and even if you are in a virtual environment it will be available everywhere. To enable swap, just type the following four lines:
```
sudo /bin/dd if=/dev/zero of=/var/swap.1 bs=1M count=1024
sudo /sbin/mkswap /var/swap.1
sudo chmod 600 /var/swap.1
sudo /sbin/swapon /var/swap.1
```

The same [Stackoverlow post](http://stackoverflow.com/questions/17173972/how-do-you-add-swap-to-an-ec2-instance) that provided this code also explains how to enable swap by default after reboot. 

If you decide you want to delete your swap (for example, if you want to create a different sized swap):
```
sudo rm /var/swap.1
```

/var/swap.1 is swap.1's absolute path. Note that this will only work if you have done just the first of the four steps above.

## Example: Running Scrapy on EC2

nb. I have included code examples here for multiple occasions on which I connected to EC2 to run this scraper, sometimes using different instances. Therefore, the coding examples are not consistent in the IP addresses used (because of the differnet occasions) or in the EC2 directory structure referenced (because of the different instances).

### 1. Launch and connect to EC2 instance

In the [EC2 console](https://console.aws.amazon.com/ec2/), launch an instance with model m4.xlarge (16 GiB memory, High Network Performance). Click on Connect and follow the instructions to connect with "A standalone SSH client" (the terminal on my macbook is a standalone SSH client). 

*Connecting to an existing EC2 isntance:* To connect to an EC2 instance that you have already set up and left running, you can just type the same code the clicking on Connect first told you to in your local terminal (and typing ```history``` in the terminal should make this even easier). It will look something like this:

```
ssh -i "/Users/sambarrows/Dropbox/AWS_keys/school-blurbs-trial-1.pem" ec2-user@ec2-54-218-235-42.us-west-2.compute.amazonaws.com
```

Where school_blurbs-trial-1.pem is the private key I saved locally when launching the instance.

Be careful with IP addresses when opening an existing (but shut down) instance. It appears that the IP address of the EC2 instance changes each time you stop and restart it. On one occassion, there was also an issue with the IP address of my local machine (I think because I am running off Harvard's network). I had to go the EC2 console, click under “Security Groups” and change the Inbound Source to My IP (see ["Adding a Rule for Inbound SSH Traffic to a Linux Instance"](http://docs.aws.amazon.com/AWSEC2/latest/UserGuide/authorizing-access-to-an-instance.html)). 

Also, rember that if you shutdown an EC2 instance you will need to enable swap again when you reopen the instance, unless you enable swat by deafult after reboot.  

### 2. Prepare the EC2 instance

There are several steps to prepare the EC2 intance:

<ol>
<li>enable swap - see above
<li>create directories - I created a directory for my virtual environemnt (Env), for my date (Data), and for my code (Code). Then within Code I made another for my spider. ***(NB. there's a much nicer way to do this and steps 3 and 4 - see LDA example below)*** 
<li>create virtual machine  - see above
<li>load packages in virtual machine
<ul>
```
pip install --upgrade pip setuptools
pip install -U numpy
pip install -U nltk
pip install -U pandas
sudo yum list gcc
sudo yum install openssl-devel
pip install scrapy
```
</ul>
</ol>
 
### 3. Change code in spider so that it reads in data from its location on EC2 

Easier to make this change locally than in a text editor on EC2.

### 4. Copy data and spider from macbook to EC2 instance

Open a separate local terminal (ie. a regular terminal on your mac) and copy your spider:

```
cd /Users/sambarrows/Dropbox/School_websites/Scraping

scp -r -i "/Users/sambarrows/Dropbox/AWS_keys/school-blurbs-trial-1.pem" blurbs_spider ec2-user@54.218.235.42:~/Scraping
```
You also need to copy the dataset holding the urls to EC2.

```
scp -r -i "/Users/sambarrows/Dropbox/AWS_keys/school-blurbs-trial-1.pem" csvurls.csv ec2-user@54.218.50.111:~/school_scraping/Data
```

### 5. Activate virtual environment, and create and log into a screen

In the EC2 terminal:
```
source /home/ec2-user/School_websites/Envs/venv/bin/activate
screen -S blurbs_scraping
```

### 5. Run Scrapy (from within virtual environment and screen)

In the EC2 terminal, first set working directory inside your spider, then:
```
scrapy crawl blurbs
```

Or to write file (*if there is an existing file by the same name, make sure to delete it first*):
```
scrapy crawl blurbs -o blurbs_dat.csv
```

Or to write file as you go:
```
scrapy crawl blurbs -t csv -o - > blurbs_dat.csv
```

Or to change settings:
```
scrapy crawl blurbs -o blurbs_dat.csv -s CONCURRENT_REQUESTS=1000 -s CONCURRENT_REQUESTS_PER_DOMAIN=1000
```

If you want to check numer of lines as it's working:
```
wc -l blurbs_dat.csv
```

### 6. When scraper has finished, copy data back to machine 

On your ***local terminal***:
```
cd <directory where you want the file to be copied> 

scp [-r] -i "<where your EC2 .pem file is>" ec2-user@54.218.235.42:~/Scraping/where/your/file/is  
``` 

***nb. when I tried this again it didn't work - it needed the destination***

Or,
```

```

For example:
```

scp -i "/Users/sambarrows/Dropbox/AWS_keys/school-blurbs-trial-1.pem" ec2-user@ec2-54-149-164-218.us-west-2.compute.amazonaws.com:/home/ec2-user/school_scraping/Code/blurbs_spider/blurbs_dat.csv /Users/sambarrows/Dropbox/School_websites/Scraping/blurbs_spider

ssh -i "/Users/sambarrows/Dropbox/AWS_keys/school-blurbs-trial-1.pem" ec2-user@ec2-54-202-156-114.us-west-2.compute.amazonaws.com
```

### 2. Create data structure you want on EC2 on your mac, and copy it across

Setting up structure on your mac saves time, and let's you run the python file locally without changing code pointing to the location of your data, to check all is working ok. Make sure that anything you are loading in your .py file points to the correct place.

Then just copy files across (***from a local terminal***, not the terminal that you just connected to EC2); assuming you just want to move it to your home directory on EC2:
```
scp -r -i "/Users/sambarrows/Dropbox/AWS_keys/school-blurbs-trial-1.pem" lda_project  ec2-user@54.202.156.114:~
```
Where 54.202.156.114 is your IPv4 Public IP, given at the bottom of your EC2 console

### 3. Prepare your EC2 instance

<ol>
<li>Enable swap - this time, I add a bit more:<br>
<ul>
```
sudo /bin/dd if=/dev/zero of=/var/swap.1 bs=1M count=4096
sudo /sbin/mkswap /var/swap.1
sudo chmod 600 /var/swap.1
sudo /sbin/swapon /var/swap.1
```
</ul><br>
I also set up the instance to enable swap every time it reboots by typing:
<ul>
```
sudo nano /etc/fstab
```
</ul><br>
And then adding the bottom line here:
<img src="swap_pic.png"><br>
<li>Create and enter virtual environment (nb. if already in location where you want the virtual environment, can give relative path after source, rather than absolute path)<br>
<ul>
```
cd Env
virtualenv lda_venv
source lda_venv/bin/activate
```
</ul><br>

One you are running the virtual environment, you will notice that the terminal has (lda_venv) at the left of the command line.<br>
<br>
<li>Load packages in virtual environment<br>
<ul>
```
pip install -U pip setuptools
pip install -U numpy
pip install -U scipy
pip install -U gensim
```
</ul><br>

<li>Create and enable screen<br>
<ul>
```screen -S blurbs_lda```
</ul>
</ol>

### 4a. Run gensim (multicore)

Then just navigate to the directory where your python file is and run it:
```
python lda_1.py
```

It returns four files: first_lda.lda is the lda model I saved. I'm not entirely sure what first_lda.lda.expElogbeta.npy, first_lda.lda.id2word, and first_lda.lda.state are.


### 4b. Run gensim (distributed)

I also tried running distributed LDA, but across multiple cores on one machine (instance), rather than multiple machines. 

At stage when install packages, need to instead install distributed version of gensim:
```
pip install -U gensim[distributed]
```

Then after you've completed all steps in (3), following the [instructions on the gensim site](https://radimrehurek.com/gensim/dist_lsi.html), set up PYRO on each machine:
```
export PYRO_SERIALIZERS_ACCEPTED=pickle
export PYRO_SERIALIZER=pickle
```

Set up several workers (if copying commands from gensim site, be sure to change lsi to lda):
```
python -m gensim.models.lda_worker &
python -m gensim.models.lda_worker &
python -m gensim.models.lda_worker &
python -m gensim.models.lda_worker &
```

Set up dispatcher:
```
python -m gensim.models.lda_dispatcher &
```

Run model:
```
python lda_1.py
```

Unfortunately, this threw up what appears to be a bug that [a lot of people have been getting](https://github.com/RaRe-Technologies/gensim/issues/911). Given that I also came across a serious bug when using gensim to find topics in school reviews that was not easy to fix, I decided to set aside using distributed gensim for now.

### 5. Copy lda model to mac

From ***local terminal***, in navigate to whichever directory you want to save the results to and then type:
```
scp -i "/Users/sambarrows/Dropbox/AWS_keys/school-blurbs-trial-1.pem" ec2-user@54.202.31.251:~/lda_project/Code/first_lda* .
```

This copies anything in code that starts with first_lda to the current directory (the . at the end tells it to copy to the current directory).

# Git

|||
|---|---|
|```git init```||
|```git status```||
|```git add .```||
|```git status```||
|```git commit -m "My initial commit"```||
|```git log```||
|```git checkout 6b8b384f43e54b7ae5b07e9f0ef20f2012bb24e3```||
